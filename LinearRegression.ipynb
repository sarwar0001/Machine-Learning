# Linear Regression


```python
import numpy as np
import pandas as pd
# matplotlib is use for visualization
import matplotlib.pyplot as plt
#sklearn use for implement the model.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Import and Print Data Set
dataset = pd.read_csv('D:/Machine_learning_practice/Salary_Data.csv')
print(dataset)
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:, 1].values
```

```python
#Split Dataset into test set and train set. Test set size is 1/3 of total size.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
#Implement Linear Regression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Predict test result
y_pred = regressor.predict(x_test)
```

```python
#Visualize Train set Result
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('Salary vs Experience(Training Set)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualize Test Set Result
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('Salary vs Experience(Test Set)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```
