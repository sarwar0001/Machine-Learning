```python
import numpy as np
import pandas as pd
# matplotlib is use for visualization
import matplotlib.pyplot as plt
#sklearn use for implement the model.
```

```python
#Import and Print Data Set
dataset = pd.read_csv('D:/Machine_learning_practice/Position_Salaries.csv')
print(dataset)
x=dataset.iloc[:, 1:2].values
y=dataset.iloc[:, 2].values
```

```python
#Implement Random Forest Regression Model
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100,random_state=0)
regressor.fit(x,y)
```

```python
#Predict new result with Random Forest Polynomial regression
y_pred=regressor.predict([[6.5]])
print(y_pred)
```

```python
#Visualising the Random Forest Regression result
x_grid=np.arange(min(x),max(x), 0.01)
x_grid=x_grid.reshape((len(x_grid), 1))
plt.scatter(x,y,color='red')
plt.plot(x,regressor.predict(x),color='blue')
plt.title('Salary vs Position')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
